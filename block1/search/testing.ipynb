```python
import matplotlib.pyplot as plt
import igraph as ig
g = ig.Graph(directed=True, n=10, edges=[[0,1],[0,2],[1,3],[1,4],[2,5],[2,6],[3,7],[3,8],[3,9]])
g.vs["label"]=[i for i in range(0,10)]
g.vs["color"] = ["white" for i in range(0,10)]
g.es["weight"]=[i for i in range(0,9)]
#g.es["label"]=["A","B","C","D","E","F","G","H","I"]
g.es["label"]=[""+str(i) for i in range(0,9)]
layout = g.layout("tree",root=[0]);
ig.plot(g, layout=layout, target="graph.svg")
```
